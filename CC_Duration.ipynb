Tratamento dos dados

```python
  # Substitui valores 0 por NaN antes de converter para datetime - Mundanças
date_columns = [
    "Planned Completion",
    "Planned Verification",
    "Planned Regulatory Closure Date",
    "Closed",
    "Actual Regulatory Approval Date",
    "Quality Verification Completion Date",
    "Approval",
    "Created"
]

for col in date_columns:
    if col in change_df.columns:
        change_df[col] = change_df[col].replace(0, pd.NA)
        change_df[col] = pd.to_datetime(change_df[col], errors='coerce', dayfirst=True)
```

```python
 # Substitui valores 0 por NaN antes de converter para datetime - Tarefas
date_columns = [
    "Data de vencimento da tarefa",
    "Data de conclusão",
    "Approval Date",
    "Created Date"
]

for col in date_columns:
    if col in task_df.columns:
        task_df[col] = task_df[col].replace(0, pd.NA)
        task_df[col] = pd.to_datetime(task_df[col], errors='coerce', dayfirst=True)
```

```python
# Renomeia colunas do change_df
change_df = change_df.rename(columns={
    "Número": "Change Number",
    "Título": "Title",
    "Fase": "Change Phase",
    "Iniciador": "Initiator",
    "Representante da qualidade": "Quality Rep"
})
```

```python
# Renomeia colunas do task_df
task_df = task_df.rename(columns={
    "Atribuído a": "Assigned To",
    "Departamento": "Department",
    "Data de vencimento da tarefa": "Task Due Date",
    "Data de conclusão": "Completion Date"
})
```

```python
# Remove linhas do change_df onde "Change Phase" é "Cancelled" ou "Initiate" - Change
change_df = change_df[~change_df["Change Phase"].isin(["Cancelled", "Initiate"])]
```

```python
# Remove linhas do task_df onde "Change Phase" é "Cancelled" ou "Initiate" - Task
task_df = task_df[~task_df["Change Phase"].isin(["Cancelled", "Initiate"])]
```

```python
# Preenche valores nulos em "Change Manager" com o valor de "Initiator" - Change
if "Change Manager" in change_df.columns and "Initiator" in change_df.columns:
    change_df["Change Manager"] = change_df["Change Manager"].fillna(change_df["Initiator"])
```

```python
# Preenche valores nulos em "Change Manager" com o valor de "Initiator" - Task
if "Change Manager Name" in task_df.columns and "Parent Author" in task_df.columns:
    task_df["Change Manager Name"] = task_df["Change Manager Name"].fillna(task_df["Parent Author"])
```

```python
#Trantando as avaliações de impacto - Change  como "No"
for col in ["Prod Impact", "Reg. Impact", "Is there artwork impact?","Site Review Council"]:
    if col in change_df.columns:
        change_df[col] = change_df[col].fillna("No")
```

```python
if 'Tier' in change_df.columns:
    change_df.loc[change_df['Tier'] == 0, 'Tier'] = "Simple"
```
